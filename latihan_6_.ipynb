```python
arr = []

# input 10 angka
for i in range(10):
    angka = int(input("Masukkan angka: "))
    arr.append(angka)

print("Array:", arr)
print("Nilai maksimum:", max(arr))
print("Nilai minimum:", min(arr))
print("Rata-rata:", sum(arr)/len(arr))
```

```python
import matplotlib.pyplot as plt
plt.bar(range(len(arr)), arr)
plt.title("Grafik Array")
plt.xlabel("Indeks")
plt.ylabel("Nilai")
plt.show()
```
