5 Nama Mahasiswa

```python
mahasiswa =  ["andi", "budi", "rehan", "gusde", "manta"]
print("List Mahasiswa : ")
print(mahasiswa)
```

Tambahan 1 Nama Baru dengan append

```python
mahasiswa.append("galang")
print("Setelah ditambahkan nama baru")
print(mahasiswa)
```

Hapus 1 elemen menggunakan remove


```python
mahasiswa.remove("galang")
print("Setelah menghapus nama galang")
print(mahasiswa)
```
